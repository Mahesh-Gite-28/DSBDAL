```python
Q1 = iris['sepal_width'].quantile(0.25)

Q3 = iris['sepal_width'].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR

upper_limit = Q3 + 1.5 * IQR

outliers = iris[(iris['sepal_width'] < lower_limit) | (iris['sepal_width'] > upper_limit)]

print(outliers)
```
