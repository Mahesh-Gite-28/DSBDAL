```python
import nltk

# Tokenization
from nltk.tokenize import sent_tokenize, word_tokenize

# Stopwords
from nltk.corpus import stopwords

# POS Tagging
from nltk import pos_tag

# Stemming
from nltk.stem import PorterStemmer

# Lemmatization
from nltk.stem import WordNetLemmatizer

# TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Download tokenizer
nltk.download('punkt')

# Download stopwords
nltk.download('stopwords')

# Download POS tagger
nltk.download('averaged_perceptron_tagger')

# Download word dictionary for lemmatization
nltk.download('wordnet')
```

```python
text = """Mahesh is learning Text Analytics using Python. He enjoys coding and studying Data Science. Natural Language Processing is very interesting."""
```

```python
print("\n========== SENTENCE TOKENIZATION ==========\n")

# Convert paragraph into sentences
sentences = sent_tokenize(text)

# Print sentences
print(sentences)
```

```python
print("\n========== WORD TOKENIZATION ==========\n")

# Convert text into words
words = word_tokenize(text)

# Print words
print(words)
```

```python
print("\n========== POS TAGGING ==========\n")

# Apply POS tagging
pos_tags = pos_tag(words)

# Print POS tags
print(pos_tags)
```

```python
print("\n========== STOPWORD REMOVAL ==========\n")

# Load English stopwords
stop_words = stopwords.words('english')

# Remove stopwords
filtered_words = [
    word for word in words
    if word.lower() not in stop_words
]

# Print filtered words
print(filtered_words)
```

```python
print("\n========== STEMMING ==========\n")

# Create stemming object
ps = PorterStemmer()

# Apply stemming
stemmed_words = [ps.stem(word) for word in filtered_words]

# Print stemmed words
print(stemmed_words)
```

```python
# Create lemmatizer object
lemmatizer = WordNetLemmatizer()

# Apply lemmatization
lemmatized_words = [
    lemmatizer.lemmatize(word)
    for word in filtered_words
]

# Print lemmatized words
print(lemmatized_words)
```

```python
# ============================================
# TERM FREQUENCY (TF)
# ============================================


# Step 1 : Document

document = "data science is important and data science is growing"


# ============================================
# STEP 2 : SPLIT INTO WORDS
# ============================================

words = document.lower().split(" ")

print("Split Words:")
print(words)


# ============================================
# STEP 3 : MAKE LIST OF UNIQUE WORDS
# ============================================

unique_words = list(set(words))

print("\nUnique Words:")
print(unique_words)


# ============================================
# STEP 4 : CREATE DICTIONARY
# ============================================

# Initially every word frequency = 0

tf = dict.fromkeys(unique_words, 0)

print("\nInitial Dictionary:")
print(tf)


# ============================================
# STEP 5 : COUNT WORD FREQUENCY
# ============================================

for word in words:

    # Increment frequency
    tf[word] += 1


# ============================================
# STEP 6 : PRINT TERM FREQUENCY
# ============================================

print("\nTerm Frequency:")
print(tf)
```

```python
# ============================================
# INVERSE DOCUMENT FREQUENCY (IDF)
# ============================================

import math


# ============================================
# STEP 1 : DOCUMENTS
# ============================================

documents = [
    "data science is important",
    "machine learning is important",
    "data science uses python"
]


# ============================================
# STEP 2 : FIND TOTAL DOCUMENTS
# ============================================

total_documents = len(documents)

print("Total Documents:")
print(total_documents)


# ============================================
# STEP 3 : SPLIT ALL DOCUMENTS INTO WORDS
# ============================================

split_documents = []

for doc in documents:

    words = doc.lower().split(" ")

    split_documents.append(words)


print("\nSplit Documents:")
print(split_documents)


# ============================================
# STEP 4 : FIND UNIQUE WORDS
# ============================================

unique_words = set()

for words in split_documents:

    unique_words.update(words)


print("\nUnique Words:")
print(unique_words)


# ============================================
# STEP 5 : CREATE IDF DICTIONARY
# ============================================

idf = dict.fromkeys(unique_words, 0)

print("\nInitial IDF Dictionary:")
print(idf)


# ============================================
# STEP 6 : COUNT IN HOW MANY DOCUMENTS
#          EACH WORD APPEARS
# ============================================

for word in unique_words:

    document_count = 0

    for doc_words in split_documents:

        if word in doc_words:

            document_count += 1


    # ========================================
    # STEP 7 : APPLY IDF FORMULA
    # ========================================

    idf[word] = math.log(total_documents / document_count)


# ============================================
# STEP 8 : PRINT IDF
# ============================================

print("\nInverse Document Frequency (IDF):")
print(idf)
```
