```python
import pandas as pd
import numpy as np
```

```python
# Load iris dataset
df = pd.read_csv("iris.csv")

# Display first 5 rows
print(df.head())
```

```python
print(df.info())
```

```python
# Group data by Species and calculate statistics
grouped_stats = df.groupby('Species').agg(
    ['mean', 'median', 'min', 'max', 'std']
)

print(grouped_stats)
```

```python
# Create lists for each species

setosa_list = df[df['Species'] == 'Iris-setosa']['SepalLengthCm'].tolist()

versicolor_list = df[df['Species'] == 'Iris-versicolor']['SepalLengthCm'].tolist()

virginica_list = df[df['Species'] == 'Iris-virginica']['SepalLengthCm'].tolist()

print("Setosa List:")
print(setosa_list)

print("\nVersicolor List:")
print(versicolor_list)

print("\nVirginica List:")
print(virginica_list)
```

```python
# Statistics for Iris-setosa

setosa = df[df['Species'] == 'Iris-setosa']

print(setosa.describe())
```

```python
# Statistics for Iris-versicolor

versicolor = df[df['Species'] == 'Iris-versicolor']

print(versicolor.describe())
```

```python
# Statistics for Iris-virginica

virginica = df[df['Species'] == 'Iris-virginica']

print(virginica.describe())
```
