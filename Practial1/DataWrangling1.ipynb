```python
import pandas as pd  #library for handling data (tables like Excel)
import numpy as np  # used for numerical operations
```

```python
# Load the Iris dataset from local CSV file into a pandas DataFrame
df = pd.read_csv("Iris.csv")
#url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"

# Display first 5 rows of the dataset to verify it loaded correctly
df.head()
```

```python
df.isnull().sum()

df.dropna(inplace=True)

#isnull() → checks if value is missing
#sum() → counts missing values
```

```python
df.describe()

#provides summary statistics like count, mean, standard deviation, minimum, maximum, and quartiles for numerical columns in the dataset.”
```

```python
df.dtypes   # shows data types of each column
```

```python
df['Species'] = df['Species'].astype('category')  # convert text to category for memory and performance optimization

#category is a datatype in pandas.
#“Category datatype is used when a column has limited repeated values. Instead of storing full strings repeatedly, 
# it stores them as numerical codes internally, which reduces memory usage and improves performance.”
```

```python
df.dtypes   # shows data types of each column
```

```python
#“Normalization is a preprocessing technique used to scale numerical data into a fixed range (usually 0 to 1) so that all features contribute equally. It is mainly used in machine learning when features have different ranges.”

numeric_columns = df.select_dtypes(include=['float64']).columns
df_normalized = df.copy()


for col in numeric_columns:
    df_normalized[col] = (df[col] - df[col].min()) / (df[col].max() - df[col].min())


print("\nFirst 5 rows after Normalization:")
df_normalized.head()


# drawback of min max normalization-->It is sensitive to outliers. A single massive outlier can skew the scaled data into a very small range.
#Purpose: Essential for machine learning algorithms that rely on distance metrics
```

```python
# 8. Convert Categorical to Quantitative (Encoding)
df_encoded = pd.get_dummies(df, columns=['Species'])  # It creates new columns for each category of Species,where each category is represented by 1 or 0, making the data suitable for machine learning algorithms.

print("\nFirst 5 rows after One-Hot Encoding:")

df_encoded.head()

#Converts categorical/text data into numerical format.
#Creates separate binary columns.

#One-Hot Encoding?: Technique that converts categories into binary columns containing 0 and 1.
```
